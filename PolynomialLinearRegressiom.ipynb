```python
# import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generating numbers between 1 to 100 and some of the outliers in it
x=6*np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
# Quadratic equation used- y=0.5x^2+1.5x+2+outliers
plt.scatter(x,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Let's implement the Simple Linear Regression
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=regression_1.predict(x_test)  
score=r2_score(y_test,y_pred)
print(score)
```

```python
# Visualise the model
plt.plot(x_train,regression_1.predict(x_train),color='r')
plt.scatter(x_train,y_train)
plt.xlabel('X_dataset')
plt.ylabel('Y_dataset')
```

```python
# Lets import polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)#This include feature helps to take first term also that is behta not
x_train_poly=poly.fit_transform(x_train)
# The main meaning of using fit_transform is that it make sures that main data validation part from where we are trying to transform is my training data no reference of test data is taken.
x_test_poly=poly.transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)
y_pred=regression.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
# Making degree value 3
poly=PolynomialFeatures(degree=2,include_bias=True)#This include feature helps to take first term also that is behta not
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)
y_pred=regression.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
# Prediction of new data
x_new=np.linspace(-3,3,200).reshape(200,1)
x_new_poly=poly.transform(x_new)
x_new_poly
```

```python
y_new=regression.predict(x_new_poly)
plt.plot(x_new,y_new,"r-",linewidth=2,label="New Predictions")
plt.plot(x_train,y_train,"b.",label='Training points')
plt.plot(x_test,y_test,"b.",label='Testing points')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

Pipelining concept

```python
# By using pipeline we will be combining the two model Polyomial Feature and Linear Regression
from sklearn.pipeline import Pipeline
```

```python

def poly_regression(degree):
    x_new=np.linspace(-3,3,200).reshape(200,1)

    poly_feature=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("poly_feature",poly_feature),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(x_train,y_train)
    y_pred_new=poly_regression.predict(x_new)
# Plotting prediction line
    plt.plot(x_new,y_pred_new,'r',label="Degree " + str(degree), linewidth=1)
    plt.plot(x_train,y_train,"b.",linewidth=3)
    plt.plot(x_test,y_test,"b.",linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
poly_regression(5)
```
